0.4 Aggregated query to select summacy rows of loans

```python
dloans = pd.read_sql_query('''select d.A2 as district_name, 
d.A3 as region,
sum(l.amount) as total_borrowed,
count(l.loan_id) as no_of_loans,
round(avg(l.amount)) as avg_loan_val 
from loan l
join account a using (account_id)
join district d 
on a.district_id = d.A1
group by d.A2, d.A3
order by sum(l.amount) DESC
limit 30''', engine)
```
